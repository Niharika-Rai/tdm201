2d. Written Explanation of each command. 

cd - Brings us back to the home directory. 
cd . - Tells us to stay in the same directory, which is the /zillow directory. 
cd .. - Takes us 1 directory up to the parent directory. In this case, it takes us to the /data directory. 
cd ../../ - Takes us 2 directories up to the parent directory. In this case, it takes us to the /tdm directory.
ls or ls . - Both give the same output, that is they display the file names in the /zillow directory. 
ls -la or ls -la . - Both give the same output, that is they display more information about the file contents in the /zillow directory. 
ls ../ - This displays the names of the subdirectories in the parent directory that is 1 level up from the current directory. Hence, this is the /data directory content names. However, the pwd command shows that our current directory is still the /zillow directory. 

2e. From my observation of the code outputs, I believe that using relative paths for particular commands like ls can benefit us by allowing more flexibility in viewing file names and contents of parent directories, while staying the same directory. For example, using just ls would give the file names of the present directory. But, using ls ../ allows us to explore the contents of the parent directory without changing your current directory.
