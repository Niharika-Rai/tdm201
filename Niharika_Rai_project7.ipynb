```python
%%sql 
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID;
```

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 7500000;
```

```python
%%sql 
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numPlayers
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID;
```

```python
%%sql 
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numPlayers
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 6800000 AND COUNT(s.playerID) > 20;
```

```python
%%sql 
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numPlayers
FROM Salaries s
JOIN Teams t ON t.teamID = s.teamID AND t.yearID = s.yearID
WHERE s.yearID BETWEEN 2010 AND 2015
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 6800000 AND COUNT(s.playerID) > 20;
```

## Question 5

```python
%%sql
WITH payroll AS (
    SELECT
        teamID,
        yearID,
        SUM(salary) AS totalPayroll
    FROM Salaries
    GROUP BY teamID, yearID
),
homeruns AS (
    SELECT
        teamID,
        yearID,
        MAX(HR) AS maxHR
    FROM Teams
    GROUP BY teamID, yearID
)
SELECT
    p.teamID,
    p.yearID,
    p.totalPayroll,
    h.maxHR
FROM payroll p
JOIN homeruns h
    ON p.teamID = h.teamID AND p.yearID = h.yearID
WHERE p.totalPayroll > 150000000 AND h.maxHR >= 30
ORDER BY p.totalPayroll ASC
LIMIT 5;
```

5.1 I used the sample query in the question and added the HR aggregration, 'JOIN', 'WHERE', 'ORDER BY' and 'LIMIT' clauses. I was confused on how to join two queries where both have been constructed using the 'WITH' clause and needed ChatGPT's assisstance. I learnt that both queries could be defined under the same 'WITH' clause but separated by a comma. Then, the 'JOIN' clause can be applied once they are aggregated. 
5.2 Based on the output for the query created in 5.1, the top 5 cheapest team-seasons that meet the given criteria are BOS, NYA, DET, CHN, and LAA.
