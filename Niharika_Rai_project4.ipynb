```python
%%bash
awk -F, '
NR>1 && $30 != ""{
    gender = ($2 == "" ? "Blank" : $2)
    vendor = $22
    rating = $30
    if (rating > max[gender]) {
        max[gender] = rating
        top_vendor[gender] = vendor
    }
}
END {
    for (g in max) {
        print g, top_vendor[g], max[g]
    }
}' /anvil/projects/tdm/data/restaurant/test_full.csv
```

```python
%%bash 
awk -F, '
NR>1 && $30 != "" {
    gender = ($2 == "" ? "Blank" : $2)
    rating = $30         
    sum[gender] += rating
    count[gender]++
}
END {
    for (g in sum) {
        avg = sum[g] / count[g]
        printf "%s %.2f\n", g, avg
    }
}' /anvil/projects/tdm/data/restaurant/test_full.csv
```
